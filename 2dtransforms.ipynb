```python
def prepare_trajectory_data( row, kind ):
    if kind in ['x', 'y', 't']:
        if row < len(file):
            ret = np.array([])
            for i in range(184):
                epoch = str(i+1)
                value = file[f"ms{epoch}_{kind}"][row]
                if 'IND' not in value:
                    ret = np.append( ret, [[float(value)]] )
            return ret
        else:
            print('your row number is too large')
    else:
        print('kind needs to be x, y or t')
    
def trajectory_dict( row ):
    return { 
        'x': prepare_trajectory_data( row, 'x' ),
        'y': prepare_trajectory_data( row, 'y' ),
        't': prepare_trajectory_data( row, 't' ),
    }

def get_trajectory( row ):
    return { 
        'x': prepare_trajectory_data( row, 'x' ) + center_point[0],
        'y': prepare_trajectory_data( row, 'y' ),
        't': prepare_trajectory_data( row, 't' ),
    }

    
    
def center_from_raw( position ):
#     return [position[0] + center_point[0], position[1] + center_point[1]]
    return [position[0] + center_point[0], position[1]]

def center_trajectory_from_raw( row_number ):
    row_data = trajectory_dict( row_number )
    ret = {
        'x': np.array([]),
        'y': row_data['y'],        
        't': row_data['t']
    }
    for index in range(len( row_data['x'] )):
        ret['x'] = np.append( ret['x'],  center_from_raw( [ row_data['x'][index], row_data['y'][index] ] )[0] ) 
    return ret
```
